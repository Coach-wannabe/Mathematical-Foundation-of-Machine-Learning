# Rental of bikes by month

# How weather affects count of rented bikes 

```python
fig = sns.boxplot(data=df, y="cnt", x="weathersit", orient="v")
fig.set(xlabel='Weather Situation', ylabel='Amount of rented bikes', title="Box Plot On Count Across Weather Situations")
```

# How, wheather customers are having a working day or not they will rent a bike  

```python
fig = sns.boxplot(data=df, y="cnt", x="workingday", orient="v")
fig.set(xlabel='Working Day', ylabel='Amount of rented bikes',
        title="Box Plot On Count Across Working Day")
```

# In what hours people tend to rent a bike

```python
fig = sns.boxplot(data=df,y="cnt",x="hr",orient="v")
fig.set(xlabel='Hour Of The Day', ylabel='Amount of rented bikes',
        title="Box Plot On Count Across Hour Of The Day")
```

# During what temperature people will and won't ride a bike

```python
sns.set(font_scale=2)
fig, axes = plt.subplots(nrows=1)
fig.set_size_inches(30, 10)
fig = sns.boxplot(data=df, y="cnt", x="temp",orient="v")
fig.set(xlabel='Temperature', ylabel='Amount of rented bikes',
        title="Box Plot On Count Across Temperature")
fig.set_xticklabels(fig.get_xticklabels(), rotation=45, ha="right")
```

# Same as for the temperature, but with wind
